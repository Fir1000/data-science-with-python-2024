```python
# Challenge 10: Import Iris dataset
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Challenge 11: Explore Iris dataset
print(iris_df.head())  # First 5 rows
print(iris_df.describe())  # Summary statistics

# Challenge 12: Plot Iris dataset
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(iris_df, hue='target', diag_kind='kde')
plt.show()

# Challenge 13: Export Iris dataset to CSV file
iris_df.to_csv('iris_dataset.csv', index=False)
print("Iris dataset exported to iris_dataset.csv")

# Challenge 14: Install yfinance (Run this in a terminal if not installed)
# !pip install yfinance

# Challenge 15: Import yfinance package
import yfinance as yf

# Challenge 17: Retrieve NVIDIA stock data
nvidia = yf.Ticker("NVDA")
nvidia_df = nvidia.history(period="1y")  # Last 1 year of data

# Challenge 18: Explore NVIDIA data
print(nvidia_df.head())  # First 5 rows
print(nvidia_df.describe())  # Summary statistics

# Challenge 19: Plot NVIDIA stock data
plt.figure(figsize=(10, 5))
plt.plot(nvidia_df.index, nvidia_df['Close'], label='NVIDIA Closing Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.title('NVIDIA Stock Price Over Last Year')
plt.legend()
plt.show()

# Challenge 20: Export NVIDIA data to CSV file
nvidia_df.to_csv('nvidia_stock.csv')
print("NVIDIA stock data exported to nvidia_stock.csv")
```

```python
# 📌 Challenge 10: Import Iris dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# 📌 Challenge 11: Explore Iris dataset  
print(iris_df.info())  # แสดงข้อมูลโดยรวม
print(iris_df.describe())  # ค่าสถิติโดยรวม
print(iris_df['species'].value_counts())  # นับจำนวนแต่ละ species  

# 📌 Challenge 12: Plot Iris dataset  
plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()

sns.pairplot(iris_df, hue="species", palette="husl")
plt.show()

# 📌 Challenge 13: Export Iris dataset  
iris_df.to_csv("iris_dataset_v2.csv", index=False)
print("✅ Exported iris dataset to 'iris_dataset_v2.csv'")

# 📌 Challenge 14: Install yfinance (หากยังไม่ได้ติดตั้งให้ใช้คำสั่งนี้)  
# !pip install yfinance  

# 📌 Challenge 15: Import yfinance  
import yfinance as yf

# 📌 Challenge 17: Retrieve NVIDIA stock data  
nvda = yf.Ticker("NVDA")
nvda_df = nvda.history(period="6mo")  # ดึงข้อมูล 6 เดือนล่าสุด

# 📌 Challenge 18: Explore NVIDIA data  
print(nvda_df.info())  # เช็คข้อมูลเบื้องต้น
print(nvda_df.describe())  # เช็คค่าสถิติโดยรวม  

# 📌 Challenge 19: Plot NVIDIA data (ปรับแต่งสีและสไตล์)  
plt.figure(figsize=(10, 5))
sns.lineplot(data=nvda_df, x=nvda_df.index, y="Close", label="Closing Price", color="green")
plt.fill_between(nvda_df.index, nvda_df["Close"], alpha=0.2, color="green")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.title("NVIDIA Stock Closing Price (Last 6 Months)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# 📌 Challenge 20: Export NVIDIA data  
nvda_df.to_csv("nvidia_stock_v2.csv", index=False)
print("✅ Exported NVIDIA stock data to 'nvidia_stock_v2.csv'")
```
